### Problem Statement<a id="0"></a>


The objective of this project is to conduct a comprehensive data analysis on the BRFSS 2021 dataset to understand the factors associated with diabetes and develop a predictive machine learning model. Specifically, this project aims to:

1. **Identify Significant Predictors:** Analyze the dataset to identify key demographic, behavioral, and health-related variables that are significantly associated with the presence of diabetes.
2. **Develop a Predictive Model:** Build and validate a machine learning model to predict the probability of diabetes diagnosis in individuals based on the identified predictors.
3. **Enhance Public Health Insights:** Utilize the findings to provide data-driven insights that can inform public health policies and strategies for diabetes prevention and management.

By achieving these objectives, the project seeks to contribute to the understanding of diabetes risk factors and support the development of effective intervention programs.



### Summary of the Dataset:<a id="2"></a>



---

- **Diabetes_binary (Target Variable)**: This binary variable indicates whether an individual has diabetes (1) or not 
- **children** : no of children.represented on a numeric scale from 1 to 10.
- **employe** : Employe level
- **HighBP**: This binary variable signifies the presence (1) or absence (0) of high blood pressure.
- **HighChol**: This variable indicates whether an individual has high cholesterol (1) or not (0).
- **CholCheck**: This binary variable shows if cholesterol has been checked (1) or not (0).
- **BMI**: Body Mass Index, which calculates body fat based on height and weight.
- **Smoker**: This binary variable indicates smoking status, with smokers represented by 1 and non-smokers by 0.
- **Stroke**: This binary variable denotes whether an individual has a history of stroke (1) or not (0).
- **HeartDiseaseorAttack**: This variable shows the presence (1) or absence (0) of heart disease or a history of heart attacks.
- **PhysActivity**: This binary variable indicates participation (1) or lack of participation (0) in physical activity.
- **Fruits**: This variable indicates regular consumption (1) or non-consumption (0) of fruits.
- **Veggies**: This variable indicates regular consumption (1) or non-consumption (0) of vegetables.
- **HvyAlcoholConsump**: This binary variable signifies heavy alcohol consumption (1) or not (0).
- **AnyHealthcare**: This variable indicates whether an individual uses (1) or does not use (0) any healthcare services.
- **NoDocbcCost**: This binary variable indicates if cost prevents doctor visits (1) or not (0).
- **GenHlth**: General health status, represented on a numeric scale from 1 to 5.
- **MentHlth**: Mental health status, represented on a numeric scale from 0 to 30.
- **PhysHlth**: Physical health status, represented on a numeric scale from 0 to 30.
- **DiffWalk**: This binary variable shows whether an individual has difficulty (1) or ease (0) in walking.
- **Sex**: This binary variable indicates gender, with males represented by 1 and females by 0.
- **Age**: Age of the survey respondents, given in years.
- **Education**: Education level of respondents, represented on a numeric scale from 1 to 6.
- **Income**: Household income level, represented on a numeric scale.

### Removing Features

Here we remove features deemed irrelevant for modeling purposes. We specify a list of columns to be removed and then drop these columns from the combined DataFrame.

```python
# Remove irrelevant features from the combined dataset
columns_to_remove = ["CholCheck", "AnyHealthcare", "NoDocbcCost",'Fruits' ,'Veggies']
df = df.drop(columns=columns_to_remove)
```

## Model Development <a id="20"></a>
---

```python
df.head()
```

```python
# convert 1/0 binary columns to boolean values
binary_columns = [
    "Diabetes",
    "HighBP",
    "HighChol",
    "Smoker",
    "Stroke",
    "HeartDiseaseorAttack",
    "PhysActivity",
    "HvyAlcoholConsump",
    "DiffWalk",
    "Sex",
]
for column in binary_columns:
    df[column] = df[column].astype("bool")
```

```python
# Selecting numerical columns (excluding binary/boolean columns)
int_columns = ["BMI", "GenHlth", "MentHlth", "PhysHlth", "Age", 'employ','Education','Income']
# Initialize the MinMaxScaler
scaler = MinMaxScaler()
# Fit and transform the numerical features
df[int_columns] = scaler.fit_transform(df[int_columns])

df.head()
```

```python
x = df.drop(columns=['Diabetes'])
y = df['Diabetes']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

print("xtrain shape: ", x_train.shape)
print("xtest shape: ", x_test.shape)
print("ytrain shape: ", y_train.shape)
print("ytest shape: ", y_test.shape)
```

```python
reg = LogisticRegression(max_iter=19500)
reg.fit(x_train, y_train)
```

```python

y_pred_test = reg.predict(x_test)

reg_accuracy_test = accuracy_score(y_test, y_pred_test)
print(f"Test Accuracy: {round(reg_accuracy_test * 100,2)}%")

y_pred_train = reg.predict(x_train)

reg_accuracy_train = accuracy_score(y_train, y_pred_train)
print(f"Train Accuracy: {round(reg_accuracy_train * 100,2)}%")
```

```python
cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d' , cmap='coolwarm')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(y_test, y_pred_test))
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)
```

```python
y_pred_knn=knn.predict(x_test)


knn_accuracy_test = accuracy_score(y_test, y_pred_knn)
print(f"Test Accuracy: {round(knn_accuracy_test * 100,2)}%")
```

```python
cm = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d' , cmap='coolwarm')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for knn')
plt.show()
```

```python
print(classification_report(y_test , y_pred_knn))
```
